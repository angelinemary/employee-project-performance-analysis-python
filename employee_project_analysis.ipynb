```python
# importing pandas library for dataframe operations
import pandas as pd

# importing numpy library for numerical operations
import numpy as np
```

```python
# creating employee dataframe with details
employee = pd.DataFrame({
    "ID": ["A001","A002","A003","A004","A005"],  # employee id
    "Name": ["John Alter","Alice Luxumberg","Tom Sabestine","Nina Adgra","Amy Johny"],  # full name
    "Gender": ["M","F","M","F","F"],  # gender
    "City": ["Paris","London","Berlin","Newyork","Madrid"],  # city
    "Age": [25,27,29,31,30]  # age
})

# displaying dataframe
employee
```

```python
# creating seniority dataframe
seniority = pd.DataFrame({
    "ID": ["A001","A002","A003","A004","A005"],  # employee id
    "Designation Level": [2,2,3,2,3]  # designation level
})

# displaying dataframe
seniority
```

```python
# creating project dataframe with project details
project = pd.DataFrame({
    "ID": ["A001","A002","A003","A004","A005","A002","A005","A003","A001","A003","A001","A004","A004","A005"],  # employee id
    "Project": ["Project1","Project2","Project3","Project4","Project5","Project6","Project7","Project8","Project9","Project10","Project11","Project12","Project13","Project14"],  # project name
    "Cost": [1002000,2000000,4500000,5500000,np.nan,680000,400000,350000,np.nan,300000,2000000,1000000,3000000,200000],  # project cost
    "Status": ["Finished","Ongoing","Finished","Ongoing","Finished","Failed","Finished","Failed","Ongoing","Finished","Failed","Ongoing","Finished","Finished"]  # project status
})

# display dataframe
project
```

```python
# saving employee dataframe to csv file
employee.to_csv("employee.csv", index=False)

# saving seniority dataframe to csv file 
seniority.to_csv("seniority.csv", index=False)

# saving project dataframe to csv file
project.to_csv("project.csv", index=False)
```

```python
# reading project csv file
project = pd.read_csv("project.csv")

# display dataframe
project
```

```python
# converting cost column into list
cost_list = project["Cost"].tolist()

# loop through list
for i in range(len(cost_list)):
    
    # check if value is missing
    if pd.isna(cost_list[i]):
        
        # calculate running average of previous values
        avg = np.nanmean(cost_list[:i])
        
        # replace missing value with average
        cost_list[i] = avg

# assign updated values back to dataframe
project["Cost"] = cost_list

# display updated dataframe
project
```

```python
# reading employee csv
employee = pd.read_csv("employee.csv")

# split name into first and last name
employee[["First Name","Last Name"]] = employee["Name"].str.split(" ", expand=True)

# remove original name column
employee.drop("Name", axis=1, inplace=True)

# display dataframe
employee
```

```python
# reading seniority csv
seniority = pd.read_csv("seniority.csv")

# merge employee and seniority
final = pd.merge(employee, seniority, on="ID")

# merge above result with project
final = pd.merge(final, project, on="ID")

# display final dataframe
final
```

```python
# giving 5% bonus if project finished
final["Bonus"] = np.where(final["Status"]=="Finished", final["Cost"]*0.05, 0)

# display dataframe
final
```

```python
# increase designation level by 1 if project failed
final.loc[final["Status"]=="Failed", "Designation Level"] += 1

# remove employees whose level greater than 4
final = final[final["Designation Level"]<=4]

# display dataframe
final
```

```python
# add Mr or Mrs based on gender
final["First Name"] = np.where(final["Gender"]=="M",
                               "Mr. " + final["First Name"],
                               "Mrs. " + final["First Name"])

# remove gender column
final.drop("Gender", axis=1, inplace=True)

# display dataframe
final
```

```python
# promote employees whose age above 29
final.loc[final["Age"]>29, "Designation Level"] -= 1

# display dataframe
final
```

```python
# calculate total cost per employee
TotalProjCost = final.groupby(["ID","First Name"])["Cost"].sum().reset_index()

# rename cost column
TotalProjCost.rename(columns={"Cost":"Total Cost"}, inplace=True)

# display dataframe
TotalProjCost
```
